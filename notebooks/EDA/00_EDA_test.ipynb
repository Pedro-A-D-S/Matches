##### Plotting ID types distribuition

```python
NA1_shape = df[df['gameid'].str.contains('NA1')].shape[0]
ESPORTSTMNT_shape = df[df['gameid'].str.contains('ESPORTSTMNT')].shape[0]
game_shape = df[df['gameid'].str.contains('_game')].shape[0]

data = {'NA': NA1_shape,
        'ESPORTSTMNT': ESPORTSTMNT_shape,
        'GAME': game_shape}

df_plot = pd.DataFrame.from_dict(data, orient = 'index', columns = ['Value'])
df_plot.reset_index(inplace=True)
df_plot.rename(columns = {'index': 'Label'}, inplace=True)

fig = px.bar(df_plot, x = 'Label', y = 'Value', template = 'plotly_dark', text_auto = '.3s')

fig.update_xaxes(categoryorder = 'total descending')

fig.update_layout(
    title = {
        'text': "Quantity of ID Types"},
    xaxis_title = 'ID Types',
    yaxis_title = 'Quantity',
    width = 1200,
    height = 500,
    font_family="Helvetica",
    font_color="White",
    font = dict(size = 18),
    title_font_family="Helvetica",
    title_font_color= "White",
    bargap = 0.4,
    uniformtext_minsize = 8)

fig.update_traces(textposition = 'outside', textfont_size = 16, marker_color = '#C70039')

fig.show()
```

# URL

The column labeled "URL" was intended to link to websites containing information about the matches. However, it is noteworthy that the majority of matches possessing the datacompleteness status of 'complete' lack a corresponding URL.

```python
num_missing = df["url"].isnull().sum()
num_urls = df["url"].notnull().sum()

print("Number of missing URLs: ", num_missing)
print("Number of URLs: ", num_urls)
```

```python
# create a DataFrame with the URL data
data_url = pd.DataFrame({
    "labels": ["Missing URLs", "Non-Missing URLs"],
    "values": [num_missing, num_urls]
})

# create the pie chart using Plotly Express with dark theme
fig = px.pie(data_url, values='values', names='labels', title='URL Availability',
             template="plotly_dark", color_discrete_sequence=px.colors.sequential.RdBu)

# show the chart
fig.show()
```

As the dataset reveals an 85.04% absence of URLs in this particular column, it may be imprudent to employ it in subsequent analyses.

# League

The column labeled "League" specifies the origin of the teams participating in the matches.

```python
df['league'].unique()
```

In total, there are fifty distinct leagues present. Let us determine the leagues that exhibit a greater volume of matches.

```python
# retrieve the initial ID within each group of duplicated IDs.
df_unique_id = df.groupby('gameid').first().reset_index()

# Group the dataset by the "league" column, calculate the total number of game IDs for each league, and arrange the result in descending order.
df_unique_id = df_unique_id.groupby('league').agg({'gameid':'count'}).sort_values(by = 'gameid',
                                                                                  ascending = False).reset_index()

# Generate a dataframe of the top 10 leagues based on the highest number of matches played.
df_top_10 = df_unique_id.head(10)
```

```python
df_top_10
```

```python
fig = px.bar(df_top_10.sort_values(by = 'gameid', ascending = True), y = 'league', x = 'gameid',
             template = 'plotly_dark', orientation = 'h',
             text_auto = '.3s', color_discrete_sequence=px.colors.sequential.RdBu)

fig.update_layout(
    title = {
        'text': "Top 10 Leagues: Most Matches Played"},
    xaxis_title = 'Number of Matches',
    yaxis_title = 'Leagues',
    width = 1200,
    height = 700,
    font_family="Helvetica",
    font_color="White",
    font = dict(size = 18),
    title_font_family="Helvetica",
    title_font_color= "White",
    bargap = 0.4,
    uniformtext_minsize = 8)


fig.show()
```

# Split

```python
df['split'].unique()
```

```python
df_split = df['split'].value_counts().reset_index()
df_split.rename(columns = {'index':  'Split',
                           'split':'matches'}, inplace = True)
```

```python
df_split['cumulative_matches'] = df_split['matches'].cumsum()
df_split
```

```python
fig = px.bar(df_split.sort_values(by = 'matches', ascending = True),
             y = 'Split', x = 'matches',
             template = 'plotly_dark', orientation = 'h',
             text_auto = '.3s', color_discrete_sequence=px.colors.sequential.RdBu)

fig.update_layout(
    title = {
        'text': "Leagues with most games played"},
    xaxis_title = 'Number of Matches',
    yaxis_title = 'Splits',
    width = 1200,
    height = 700,
    font_family="Helvetica",
    font_color="White",
    font = dict(size = 18),
    title_font_family="Helvetica",
    title_font_color= "White",
    bargap = 0.4,
    uniformtext_minsize = 8)


fig.show()
```

The most played matches are on Summer and Spring and there are 240 matches played with the split flag called 2022.
